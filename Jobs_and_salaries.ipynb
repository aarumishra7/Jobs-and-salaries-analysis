```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('jobs_in_data.csv')

# EDA for work_year
work_year_distribution = data['work_year'].value_counts().sort_index()

# Visualization
plt.figure(figsize=(10, 6))
work_year_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Work Years')
plt.xlabel('Work Year')
plt.ylabel('Count')
plt.show()
```

```python
# EDA for job_title and salary
job_title_salary_distribution = data.groupby('job_title')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(12, 25))
job_title_salary_distribution.plot(kind='barh', color='salmon')
plt.title('Salary Distribution Across Job Titles')
plt.xlabel('Median Salary in USD')
plt.ylabel('Job Title')
plt.show()
```

```python
# EDA for employee_residence and salary
salary_by_residence = data.groupby('employee_residence')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(14, 8))
salary_by_residence.plot(kind='bar', color='green')
plt.title('Geographical Salary Differences')
plt.xlabel('Country of Residence')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
salary_by_experience = data.groupby('experience_level')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
salary_by_experience.plot(kind='bar', color='purple')
plt.title('Salary Distribution Across Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for employment_type and salary
salary_by_employment_type = data.groupby('employment_type')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
salary_by_employment_type.plot(kind='bar', color='orange')
plt.title('Salary Distribution Across Employment Types')
plt.xlabel('Employment Type')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for work_setting and salary
salary_by_work_setting = data.groupby('work_setting')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
salary_by_work_setting.plot(kind='bar', color='blue')
plt.title('Salary Distribution Across Work Settings')
plt.xlabel('Work Setting')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for company_location and salary
salary_by_company_location = data.groupby('company_location')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(14, 8))
salary_by_company_location.plot(kind='bar', color='red')
plt.title('Salary Distribution Across Company Locations')
plt.xlabel('Company Location')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for company_size and salary
salary_by_company_size = data.groupby('company_size')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
salary_by_company_size.plot(kind='bar', color='brown')
plt.title('Salary Distribution Across Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for salary_currency and salary
salary_by_currency = data.groupby('salary_currency')['salary_in_usd'].median().sort_values(ascending=False)

# Visualization
plt.figure(figsize=(10, 6))
salary_by_currency.plot(kind='bar', color='gold')
plt.title('Salary Distribution Across Currencies')
plt.xlabel('Salary Currency')
plt.ylabel('Median Salary in USD')
plt.show()
```

```python
# EDA for job_category, work_setting, and salary
salary_by_category_and_setting = data.groupby(['job_category', 'work_setting'])['salary_in_usd'].median().unstack()

# Visualization
plt.figure(figsize=(14, 8))
salary_by_category_and_setting.plot(kind='bar', colormap='viridis')
plt.title('Salary Variation Across Job Categories and Work Settings')
plt.xlabel('Job Category')
plt.ylabel('Median Salary in USD')
plt.legend(title='Work Setting', loc='upper right')
plt.show()
```

```python
# EDA for job_category, company_size, and salary
salary_by_category_and_size = data.groupby(['job_category', 'company_size'])['salary_in_usd'].median().unstack()

# Visualization
plt.figure(figsize=(14, 8))
salary_by_category_and_size.plot(kind='bar', colormap='coolwarm')
plt.title('Salary Variation Across Job Categories and Company Sizes')
plt.xlabel('Job Category')
plt.ylabel('Median Salary in USD')
plt.legend(title='Company Size', loc='upper right')
plt.show()
```

```python
# EDA for job_category, company_location, and salary
salary_by_category_and_location = data.groupby(['job_category', 'company_location'])['salary_in_usd'].median().unstack()

# Visualization
plt.figure(figsize=(16, 10))
salary_by_category_and_location.plot(kind='bar', stacked=True, colormap='Set3')
plt.title('Salary Variation Across Job Categories and Company Locations')
plt.xlabel('Job Category')
plt.ylabel('Median Salary in USD')
plt.legend(title='Company Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# EDA for work_year, job_category, and salary
salary_by_year_and_category = data.groupby(['work_year', 'job_category'])['salary_in_usd'].median().unstack()

# Visualization using a line chart
plt.figure(figsize=(14, 8))
for category in salary_by_year_and_category.columns:
    plt.plot(salary_by_year_and_category.index, salary_by_year_and_category[category], label=category)

plt.title('Median Salary Trend Over the Years for Job Categories')
plt.xlabel('Work Year')
plt.ylabel('Median Salary in USD')
plt.legend(title='Job Category', loc='upper left')
plt.show()
```
